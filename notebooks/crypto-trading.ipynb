```python
def preprocess_data_parquet(start_timestamp, assets):
    assets_price = defaultdict(dict)
    file_names = os.listdir(f'{BASIC_PATH}/assets')
    for file_name in file_names:
        if not file_name.endswith('.parquet'):
            continue
        asset1, asset2 = file_name.rstrip('.parquet').split('-')
        if asset1 not in assets or asset2 not in assets:
            continue
        print(f'Processing {file_name}...')
        d = pd.read_parquet(f'{BASIC_PATH}/assets/{file_name}', engine='pyarrow')
        d.index = list(map(lambda x: int(datetime.datetime.timestamp(x)), d.index))
        d = d[d.index >= start_timestamp]
        d = d[~d.index.duplicated(keep='last')]
        assets_price[asset1] = {**assets_price[asset1], asset2: d['open']}
        assets_price[asset2] = {**assets_price[asset2], asset1: 1 / d['open']}
        print(f'Processed {file_name}!')
    return assets_price
```

```python
def preprocess_data_csv(start_timestamp, assets):
    assets_price = defaultdict(dict)
    file_names = os.listdir(f'{BASIC_PATH}')
    for file_name in file_names:
        if not file_name.endswith('.csv'):
            continue
        asset1, asset2 = file_name.rstrip('.csv').split('-')
        if asset1 not in assets or asset2 not in assets:
            continue
        print(f'Processing {file_name}...')
        d = pd.read_csv(f'{BASIC_PATH}/{file_name}', index_col='time')
        d.index = list(
            map(lambda x: int(datetime.datetime.timestamp(datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))),
                d.index))
        d = d[d.index >= start_timestamp]
        d = d[~d.index.duplicated(keep='last')]
        assets_price[asset1] = {**assets_price[asset1], asset2: d['open']}
        assets_price[asset2] = {**assets_price[asset2], asset1: 1 / d['open']}
        print(f'Processed {file_name}!')
    return assets_price
```

```python
AVG_BLOCK_TIME = {
    'bitcoin': 10 * 60,
    'ethereum': 15,
    'binance_smart_chain': 3,
    'doge': 60,
    'litecoin': 2.5 * 60,
    'tron': 3
}
```

```python
AVG_FEES = {
    'relative': {
        'binance-fee': 0.001,
        'coinbase-fee': 0.001,

        # withdrawal
        'coinbase-withdrawal-fee': 0.001,
    },
    'absolute': {
        'bitcoin-fee-usd': 0.957,
        'ethereum-fee-usd': 1,
        'litecoin-fee-usd': 0.03,
        'binance_smart_chain-fee-usd': 0.25,
        'doge-fee-usd': 0.1,
        'tron-fee-usd': 0,  # *has some nuances

        # withdrawals
        'ethereum-withdrawal-fee': 0.0012,
        'bitcoin-withdrawal-fee': 0.0002,
        'litecoin-withdrawal-fee': 0.001,
        'tron-withdrawal-fee': 1,
    }
}
```

```python
# Assets price:
def exchanger(timestamp, base_asset, quote_asset):
    """
    function to return the price of the quote asset in the terms of a quote asset
    for the given timestamp
    
    params:
    timestamp: int
    base_asset: str anyof[USDT, EUR, BTC]
    quote_asset: str anyof[USDT, EUR, BTC]
    return: float
    """
    if base_asset == quote_asset:
        return 1
    if base_asset not in assets_price or quote_asset not in assets_price[base_asset] or timestamp not in \
            assets_price[base_asset][quote_asset]:
        return np.inf
    return assets_price[base_asset][quote_asset][timestamp]
```

```python
def get_blockchain(asset):
    return {
        'BTC': 'bitcoin',
        'ETH': 'ethereum',
        'BNB': 'binance_smart_chain',
        'DOGE': 'doge',
        'LTC': 'litecoin',
        'USDT': 'tron',
        'RUB': 'binance_smart_chain',
        'UAH': 'binance_smart_chain',
    }[asset]


def get_blocktime(blockchain):
    return AVG_BLOCK_TIME[blockchain]


def calc_fees(timestamp, base_asset, quote_asset, base_platform, quote_platform, price):
    if base_asset != quote_asset and base_platform != quote_platform:
        raise Exception('failed to exchange!')

    if base_platform == quote_platform:
        return (
            price * AVG_FEES['relative'][f'{base_platform}-fee'],  #platform fee
            exchanger(timestamp, 'USDT', base_asset) * AVG_FEES['absolute'][f'{get_blockchain(base_asset)}-fee-usd'],
            # sender fee
            exchanger(timestamp, 'USDT', quote_asset) * AVG_FEES['absolute'][f'{get_blockchain(quote_asset)}-fee-usd']
            # receiver fee
        )

    return (
        price * AVG_FEES['relative'].get(f'{base_platform}-fee', 1) + price * AVG_FEES['relative'].get(
            f'{quote_platform}-fee', 1),
        #platform fee
        exchanger(timestamp, 'USDT', base_asset) * AVG_FEES['absolute'][f'{get_blockchain(base_asset)}-fee-usd'] +
        AVG_FEES['absolute'][f'{get_blockchain(base_asset)}-withdrawal-fee'],
        # sender fee
        exchanger(timestamp, 'USDT', quote_asset) * AVG_FEES['absolute'][f'{get_blockchain(quote_asset)}-fee-usd']
        # receiver fee
    )


def calc_transaction(timestamp, base_asset, quote_asset, base_platform, quote_platform, funds):
    price = funds * exchanger(timestamp, base_asset, quote_asset)
    platf_fees, sender_fees, receiver_fees = calc_fees(timestamp, base_asset, quote_asset, base_platform,
                                                       quote_platform, price)
    return price + platf_fees + sender_fees + receiver_fees
```

```python
def conform_timestamp(old, new):
    if (new - old) % 60 != 0:
        return new + 60 - ((new - old) % 60)
    return new
```

# Strategies:
__Note:__ _first trading greedy, without paying attention to the change in price due to transaction delay_
* Try having multiple assets in the portfolio
* Find 3 assets that could form a triangular exchange, and utilize that
* Basic loop which starts and ends in USDT during the same operation
* Jump off the crypto asset (e.g. BTC, ETH etc) only at the end of the cycle (e.g. after one week)

```python
def dry_run(t, assets_chain, trade):
    assert len(assets_chain) >= 3

    target, target_platform = assets_chain[0]

    prev_coin = target
    prev_platform = target_platform
    intermediate = trade
    for i in range(1, len(assets_chain)):
        curr_coin, curr_platform = assets_chain[i]
        intermediate = calc_transaction(t, prev_coin, curr_coin, prev_platform, curr_platform, intermediate)
        prev_coin = curr_coin

    result = calc_transaction(t, prev_coin, target, prev_platform, target_platform, intermediate)
    return result


def run(t, assets_chain, trade):
    assert len(assets_chain) >= 3

    target, target_platform = assets_chain[0]

    prev_coin = target
    prev_platform = target_platform
    intermediate = trade

    times_history = []
    for i in range(1, len(assets_chain)):
        curr_coin, curr_platform = assets_chain[i]
        intermediate = calc_transaction(t, prev_coin, curr_coin, prev_platform, curr_platform, intermediate)
        prev_coin = curr_coin
        t = conform_timestamp(
            t,
            t + get_blocktime(get_blockchain(prev_coin)) + get_blocktime(get_blockchain(curr_coin))
        )
        times_history.append(t)

    result = calc_transaction(t, prev_coin, target, prev_platform, target_platform, intermediate)
    return result, t, times_history
```

```python
def simulate(fn):
    def wrapper(portfolio, timestamps, *args, **kwargs):
        WEEK = 604800
        DAY = WEEK // 7

        report = {**{
            'coins': [],
            'dates': [],
            'weeks': [],
        }, **{k: [] for k in portfolio.keys()}, **{f'{k}_change': [] for k in portfolio.keys()}}

        min_timestamp = min(timestamps)
        timestamp = min_timestamp
        max_timestamp = max(timestamps)

        week_start = portfolio

        last_week_timestamp = min_timestamp
        last_day_timestamp = min_timestamp
        while timestamp < max_timestamp:
            next_timestamp = timestamp + 60
            next_timestamp, max_benefit, coins = fn(next_timestamp, portfolio, *args, **kwargs)

            ### Report formation:
            if max_benefit > 0:
                report['triplets'].append('-'.join(coins))
                report['weeks'].append((timestamp - min_timestamp) // WEEK)
                report['dates'].append(datetime.datetime.fromtimestamp(timestamp))
                for k in portfolio.keys():
                    report[k].append(portfolio[k])
                    report[f'{k}_change'].append(portfolio[k] - week_start[k])

            if (timestamp - last_day_timestamp) // DAY >= 1:
                print('Day:', (timestamp - min_timestamp) // DAY)
                last_day_timestamp = timestamp

            if (timestamp - last_week_timestamp) // WEEK >= 1:
                print(30 * '*')
                print(f'Week {(timestamp - min_timestamp) // WEEK} start:')
                for asset, value in week_start.items():
                    print('>', asset, ':', value)
                print('Week end:')
                for asset, value in portfolio.items():
                    print('>', asset, ':', value)

                print('Benefit:')
                for asset, value in portfolio.items():
                    print('>', asset, ':', float(portfolio[asset]) - week_start[asset])
                week_start = portfolio
                print(30 * '*')
                last_week_timestamp = timestamp

            timestamp = next_timestamp
        return report

    return wrapper
```

```python
@simulate
def strategy_simple_triangle(timestamp, portfolio, assets, max_trade_ratio, min_spread=0.015):
    max_benefit = 0

    best_trade = None

    best_start = None
    best_interm = None
    best_benefiter = None

    for start, value in portfolio.items():
        trade = value * max_trade_ratio

        for intermediater, benefiter in combinations(filter(lambda x: x != start, assets), 2):
            result = dry_run(timestamp, [start, intermediater, benefiter], trade)

            if trade * min_spread <= trade - result > max_benefit:
                max_benefit = trade - result

                best_trade = trade

                best_start = start
                best_interm = intermediater
                best_benefiter = benefiter

    if max_benefit > 0:
        result, timestamp, [time_after_first, time_after_second] = run(timestamp,
                                                                       [best_start, best_interm, best_benefiter],
                                                                       best_trade)

        print(datetime.datetime.fromtimestamp(timestamp), best_start, best_interm, best_benefiter)
        print('Trade', best_trade)
        print(f'>>> {best_start}-{best_interm}: {exchanger(timestamp, best_start, best_interm)}')
        print(f'>>> {best_interm}-{best_benefiter}: {exchanger(time_after_first, best_interm, best_benefiter)}')
        print(f'>>> {best_benefiter}-{best_start}: {exchanger(time_after_second, best_benefiter, best_start)}')
        print('Benefit', best_trade - result)

        portfolio[best_start] += best_trade - result
    return timestamp, max_benefit, [best_start, best_interm, best_benefiter]
```

```python
@simulate
def strategy_between_platforms(timestamp, portfolio, assets, max_trade_ratio, min_spread=0.005):
    max_benefit = 0

    best_trade = None

    best_from = None
    best_to = None

    for platform1, wallet1 in portfolio.items():
        for platform2, wallet2 in portfolio.items():
            if platform1 == platform2:
                continue

            for assetFrom in assets:
                for assetTo in assets:
                    if assetFrom == assetTo:
                        continue
                    if assetFrom in wallet1:
                        if wallet1[assetFrom] < wallet2[assetFrom]:
                            trade = wallet2[assetFrom] * max_trade_ratio
                            platformFrom = platform2
                            platformTo = platform1
                        elif wallet1[assetFrom] > wallet2[assetFrom]:
                            trade = wallet1[assetFrom] * max_trade_ratio
                            platformFrom = platform1
                            platformTo = platform2
                        else:
                            continue

                        result = dry_run(timestamp,
                                         [(assetFrom, platformFrom),
                                          (assetTo, platformFrom),
                                          (assetTo, platformTo),
                                          (assetFrom, platformTo)], trade)

                        if trade * min_spread <= trade - result > max_benefit:
                            max_benefit = trade - result

                            best_trade = trade

                            best_from = (assetFrom, platformFrom)
                            best_to = (assetTo, platformTo)

    if max_benefit > 0:
        result, timestamp, _ = run(timestamp,
                                   [best_from,
                                    (best_to[0], best_from[1]),
                                    best_to,
                                    (best_from[0], best_to[1])],
                                   best_trade)

        print(datetime.datetime.fromtimestamp(timestamp), best_from, best_to)
        print('Trade', best_trade)
        print(f'>>> {best_from}: {exchanger(timestamp, best_from, best_to)}')
        print('Benefit', best_trade - result)

        portfolio[best_from] += best_trade - result
    return timestamp, max_benefit, [best_from, best_to]
```
