```python
from Historic_Crypto import HistoricalData, Cryptocurrencies
import warnings
from itertools import combinations

warnings.filterwarnings("ignore")
```

```python
pairs = ['BTC', 'USDT', 'ETH', 'BNB', 'EUR', 'LTC']
```

```python
data = Cryptocurrencies(extended_output=False).find_crypto_pairs()
data
```

```python
assets_prices = {}
for p in combinations(pairs, 2):
    if f'{p[0]}-{p[1]}' in list(data.id):
        pair = f'{p[0]}-{p[1]}'
    elif f'{p[1]}-{p[0]}' in list(data.id):
        pair = f'{p[1]}-{p[0]}'
    else:
        print('Asset pair not found!', f'{p[0]}-{p[1]}')
        continue
    print(('*' * 30) + pair + ('*' * 30))
    assets_prices[pair] = HistoricalData(f'{pair}',60,'2022-01-01-00-00').retrieve_data()
assets_prices
```

```python
for name, asset in assets_prices.items():
    asset.to_csv(f'../data/coinbase/{name}.csv', index=True)
```
